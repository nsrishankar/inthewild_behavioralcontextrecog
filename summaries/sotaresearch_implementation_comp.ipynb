**Notes**
1. 0.70-0.30 train-test split
2. Five-fold cross-validation: Training set (48 users), test set (12 users)
3. Leave-one-user-out setup
4. For MLP: Input layer is 175 features from 6 sensors. Gradient descent done over 40 epochs with minibatch of 300. Learning rate was linearly decreasing at every epoch, from 0.1 to 0.01. Momentum with weight 0.5.
